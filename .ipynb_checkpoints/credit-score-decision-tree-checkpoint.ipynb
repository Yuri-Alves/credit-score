```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

# 1) Comece carregando as bases de treino (X e y) e teste (X e y).
Verifique se o número de linhas condiz, se as variáveis estão corretas sendo apenas a de score para y e as demais nas bases de X e por último, se Y está balanceada no teste.

```python
#seu código aqui
dfX_test = pd.read_csv("X_test_balanced.csv")
dfX_train = pd.read_csv("X_train_balanced.csv")
dfy_test = pd.read_csv("y_test_balanced.csv")
dfy_train = pd.read_csv("y_train_balanced.csv")
```

# 2) Explique com suas palavras, qual o passo a passo para a aplicação do algoritmo da árvore de decisão, não esqueça de citar a etapa de avaliação do modelo e também como podemos melhorar nosso modelo.

```python
#Sua resposta aqui
#Primeiro é necessário realizar o pré-processamento dos dados, lidando com valores nulos, outliers e dados categóricos, logo após é necessário saber oque será o foco do modelo, então é recomendado saber a correlação entre as variáveis que existem no DF em questão. Logo após é necessário realizar a separação das bases de treino e teste do modelo, depois de separadas elas precisam ser carregadas para que sejam usadas pelo modelo. Depois de carregadas, pode escolher qual modelo aplicar como naive bayles ou de arvore de teste, e depois disso para saber a eficácia do modelo é recomendado puxar atributos como accuracy e recall, e se necessário criar a matriz de confusão para saber a acertividade de seu modelo, assim conseguindo saber sua eficácia.
```

# 3) Aplique o algortimo da árvore de decisão aos dados de treinamento, utilizando critério de Gini e random state = 0.
Traga a acurácia para o modedlo com os dados de treino.

```python
#seu código aqui
arvore_score = DecisionTreeClassifier (criterion = 'gini', random_state = 0)
arvore_score.fit(dfX_train, dfy_train)
previsoes = arvore_score.predict(dfX_train)
acuracia = accuracy_score(dfy_train, previsoes)
print("Acurácia:", acuracia)
```

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

```python
#seu código aqui
previsoes_test = arvore_score.predict(dfX_test)
acuracia_test = accuracy_score(dfy_test, previsoes_test)
print("Acurácia:", acuracia_test)

#Tanto na base treino quanto na base de test o modelo apresenta um acerto um tanto quanto esquisito, 100% é realmente inesperado para um modelo, porém o modelo parece ser acertivo sempre.
```

# 5) Plote a árvore de decisão.
É possível fazer uma avaliação visual? Qual a profundidade da árvore?

```python
#seu código aqui
print(arvore_score.get_params())
```

```python
print("Profundidade da árvore:", arvore_score.get_depth())
```

```python
relatorio_test = classification_report(dfy_test, previsoes_test)
print("Relatório de Classificação:")
print(relatorio_test)
```

```python
matriz_confusao = confusion_matrix(dfy_test, previsoes_test)

# Plotando a matriz de confusão como um mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_confusao, annot=True, cmap='Blues', fmt='d',
            xticklabels=['Score Baixo', 'Score Medio','Score Alto'],
            yticklabels=['Score Baixo', 'Score Medio','Score Alto'])
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```

# 6) Identifique as 2 principais features do modelo.


```python
#seu código aqui
importancias_features = arvore_score.feature_importances_
nomes_features = dfX_train.columns
plt.figure(figsize=(10, 6))
plt.barh(nomes_features, importancias_features)
plt.xlabel('Importância das Features')
plt.ylabel('Features')
plt.title('Importância das Features na Árvore de Decisão')
plt.show()

#Principais features - 1= Income, 2= Home Ownership_rented
```

```python
#seu código aqui
dfX_train_reduzido = dfX_train[['Income','Home Ownership_Rented']]
dfX_test_reduzido = dfX_test[['Income','Home Ownership_Rented']]
```

```python
arvore_score_reduzido = DecisionTreeClassifier (criterion = 'gini', random_state = 0)
arvore_score_reduzido.fit(dfX_train_reduzido, dfy_train)
```
